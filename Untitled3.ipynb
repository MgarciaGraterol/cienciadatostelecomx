```python
import pandas as pd

# Cargar el dataset limpio
df = pd.read_csv("TelecomX_Data_Limpio.csv")
```

```python
df.columns
```

```python
# 2️⃣ Mostrar columnas disponibles
print("Columnas en el dataset:")
print(df.columns)

# 3️⃣ Definir columnas a eliminar
# No eliminamos 'Churn' porque es la variable objetivo
columnas_a_eliminar = ['customerID']  # Solo el ID, ya que otras pueden ser útiles para el análisis

# 4️⃣ Eliminar columnas irrelevantes
df_modelo = df.drop(columns=columnas_a_eliminar)

# 5️⃣ Revisar el nuevo dataframe
print("\nColumnas después de eliminar identificadores irrelevantes:")
print(df_modelo.columns)

# 6️⃣ Vista rápida del dataframe listo para el análisis/modelado
df_modelo.head()
```

```python
# Variables categóricas detectadas (incluyen 'Churn' que es variable objetivo)
categoricas = df_modelo.select_dtypes(include=['object']).columns.tolist()

# Quitamos 'Churn' de la lista porque la manejamos aparte
if 'Churn' in categoricas:
    categoricas.remove('Churn')

print("Variables categóricas a codificar (sin 'Churn'):")
print(categoricas)

# Separar características y variable objetivo
X = df_modelo.drop(columns=['Churn'])
y = df_modelo['Churn']

# Codificar variables categóricas con One-Hot Encoding
X_encoded = pd.get_dummies(X, columns=categoricas, drop_first=True)

# Codificar variable objetivo
y_encoded = y.map({'No': 0, 'Yes': 1})

# Resultados
print("\nColumnas después de codificación:")
print(X_encoded.columns)

print("\nEjemplo de datos codificados:")
display(X_encoded.head())

print("\nDistribución variable objetivo (Churn):")
print(y_encoded.value_counts())
```

```python
# Asumiendo que y_encoded tiene la variable objetivo binaria (0 = No, 1 = Yes)
proporciones = y_encoded.value_counts(normalize=True) * 100

print("Proporción de clientes según estado de Churn (%):")
print(proporciones)
```

```python
# Verificar si existen valores nulos en y_encoded
print("Valores nulos en y_encoded:", y_encoded.isnull().sum())

# Si hay NaN, los eliminamos junto con sus filas correspondientes en X_encoded
mask = y_encoded.notnull()
X_encoded_clean = X_encoded[mask]
y_encoded_clean = y_encoded[mask]

print("Nuevos tamaños tras eliminar NaNs:")
print("X_encoded_clean:", X_encoded_clean.shape)
print("y_encoded_clean:", y_encoded_clean.shape)
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_encoded_clean, y_encoded_clean)

print("Distribución después de undersampling:")
print(y_resampled.value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Juntamos las variables numéricas con la variable objetivo 'Churn'
# Tomamos X_encoded_clean para datos limpios y y_encoded_clean para variable objetivo
df_corr = X_encoded_clean.copy()
df_corr['Churn'] = y_encoded_clean
```

```python
# 2. Calculamos la matriz de correlación (solo variables numéricas)
corr_matrix = df_corr.corr()

# 3. Visualizamos el mapa de calor
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, square=True, cbar_kws={"shrink": .8})
plt.title("Matriz de Correlación - Variables Numéricas y Churn")
plt.show()
```

```python
# 4. Extraemos las correlaciones absolutas de cada variable con Churn y las ordenamos
corr_churn = corr_matrix['Churn'].drop('Churn').abs().sort_values(ascending=False)
print("Variables con mayor correlación absoluta con Churn:")
print(corr_churn.head(10))
```

```python
# Variables de interés
variables = ['tenure', 'Charges.Total']  # Ajusta si el nombre exacto es distinto (ejemplo: 'TotalCharges')

# Verificamos que existan las columnas y que Churn esté disponible
for var in variables:
    if var not in df_modelo.columns:
        print(f"⚠️ La variable '{var}' no está en el dataframe.")
        variables.remove(var)

if 'Churn' not in df_modelo.columns:
    print("⚠️ La variable 'Churn' no está en el dataframe.")
else:
    sns.set_style("whitegrid")
    for var in variables:
        plt.figure(figsize=(8,5))
        sns.boxplot(x='Churn', y=var, data=df_modelo, palette={"No": "#4CAF50", "Yes": "#F44336"})
        plt.title(f"Distribución de {var} según estado de cancelación (Churn)")
        plt.xlabel("Churn")
        plt.ylabel(var)
        plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Suponiendo que X_resampled y y_resampled son los datos balanceados después de undersampling
# Si quieres usar datos sin balancear, cambia por X_encoded_clean y y_encoded_clean

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled,
    test_size=0.3,      # 30% test, 70% train
    random_state=42,    # para reproducibilidad
    stratify=y_resampled  # mantiene la proporción de clases en ambos conjuntos
)

print(f"Tamaño conjunto de entrenamiento: {X_train.shape[0]} muestras")
print(f"Tamaño conjunto de prueba: {X_test.shape[0]} muestras")

print("\nDistribución clases en entrenamiento:")
print(y_train.value_counts(normalize=True))

print("\nDistribución clases en prueba:")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# 1️⃣ Instanciar modelos
dt_model = DecisionTreeClassifier(random_state=42)
rf_model = RandomForestClassifier(random_state=42, n_estimators=100)
```

```python
# 2️⃣ Entrenar modelos
dt_model.fit(X_train, y_train)
rf_model.fit(X_train, y_train)
```

```python
# 3️⃣ Predecir sobre el conjunto de prueba
y_pred_dt = dt_model.predict(X_test)
y_pred_rf = rf_model.predict(X_test)
```

```python
# 4️⃣ Evaluar modelos
print("=== Árbol de Decisión ===")
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_dt))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_dt))

print("\n=== Random Forest ===")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Reporte de clasificación:\n", classification_report(y_test, y_pred_rf))
print("Matriz de confusión:\n", confusion_matrix(y_test, y_pred_rf))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

def evaluar_modelo(nombre, y_true, y_pred):
    print(f"--- Evaluación del modelo: {nombre} ---")
    print(f"Exactitud (Accuracy): {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precisión: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred):.4f}")
    print(f"F1-score: {f1_score(y_true, y_pred):.4f}")
    print("Matriz de confusión:")
    print(confusion_matrix(y_true, y_pred))
    print("\nReporte de clasificación:")
    print(classification_report(y_true, y_pred))
    print("---------------------------------------------------\n")

# Evaluar Árbol de Decisión
evaluar_modelo("Árbol de Decisión", y_test, y_pred_dt)

# Evaluar Random Forest
evaluar_modelo("Random Forest", y_test, y_pred_rf)
```

Informe del Análisis de Cancelación de Clientes (Churn) - TelecomX
Introducción
El objetivo de este análisis es identificar los factores que influyen en la cancelación de clientes (churn) en TelecomX, y construir modelos predictivos que permitan anticipar esta cancelación para diseñar estrategias efectivas de retención.

Limpieza y Preparación de Datos
Se partió de un dataset limpio donde se eliminaron columnas irrelevantes como identificadores únicos. Las variables categóricas se codificaron con One-Hot Encoding, y la variable objetivo Churn se transformó a formato binario (0 = No, 1 = Sí).

Para balancear el conjunto, se aplicó undersampling a la clase mayoritaria para evitar sesgo en los modelos.

Modelos Construidos y Resultados
Se entrenaron dos modelos que no requieren normalización:

Árbol de Decisión

Random Forest

Métrica	Árbol de Decisión	Random Forest
Accuracy	69.5%	76.6%
Precision	69.0%	75.5%
Recall	70.9%	78.6%
F1-score	69.9%	77.0%

El Random Forest superó al Árbol de Decisión en todas las métricas, demostrando mejor capacidad para generalizar y predecir la cancelación.

Factores que Más Influyen en la Cancelación
Basado en el análisis de importancia de variables del modelo Random Forest y en la matriz de correlación, los principales factores asociados a la cancelación son:

Tiempo de contrato (tenure): Clientes con menor antigüedad tienden a cancelar más.

Tipo de contrato: Contratos a corto plazo o mes a mes tienen mayor probabilidad de churn que contratos anuales.

Método de pago: Clientes con ciertos métodos de pago (como pago manual o sin automatización) presentan mayor cancelación.

Servicios contratados: Ausencia de servicios adicionales como soporte técnico, seguridad en línea o streaming se asocian con mayor churn.

Facturación mensual: Valores altos o inconsistentes pueden motivar cancelación.

Propuestas de Estrategias de Retención
Fidelización de clientes nuevos:
Implementar programas especiales para clientes en sus primeros meses para aumentar la retención temprana.

Promoción de contratos a largo plazo:
Incentivar a los clientes para que firmen contratos anuales ofreciendo descuentos o beneficios exclusivos.

Mejora y promoción de servicios adicionales:
Promover el uso de servicios que agregan valor (soporte, seguridad, streaming) con paquetes atractivos.

Facilitar métodos de pago automáticos:
Incentivar pagos automáticos para evitar cancelaciones relacionadas con dificultades en la facturación.

Monitoreo de facturación y comunicación proactiva:
Detectar cambios abruptos en cargos mensuales y contactar preventivamente a clientes para aclarar dudas y ofrecer soluciones.
